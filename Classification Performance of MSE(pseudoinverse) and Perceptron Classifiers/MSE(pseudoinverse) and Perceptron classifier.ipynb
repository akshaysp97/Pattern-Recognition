# Perceptron Classifier

```python
#Perceptron classifier

from csv import reader
import numpy as np
import copy
import sklearn
from sklearn.preprocessing import StandardScaler
from sklearn.linear_model import Perceptron
from sklearn.metrics import accuracy_score

def ld_csv(filename):
    with open(filename, 'r') as file:
        csv_readr = reader(file, delimiter=',' )
        thelist = list(csv_readr)
    return thelist

filename = 'wine_train.csv'
dataset = ld_csv(filename)
features_train = copy.deepcopy(dataset)
for row in features_train:
    del row[-1]
train_data = np.array(features_train).astype("float")
label_train = []
for row in dataset:
    label_train.append(row[-1])    
class_label = np.array(label_train).astype("float")
print("For the unnormalized training data")
print("Mean = ", np.mean(train_data,axis = 0))
print("Standard deviation = ", np.std(train_data, axis = 0))


#Perceptron classifier for first two features
#Training data
scaler = StandardScaler()
scaler.fit(train_data)
normal_train_data = scaler.transform(train_data)
col1 = np.zeros(len(normal_train_data))
col2 = np.zeros(len(normal_train_data))
for i in range(len(normal_train_data)):
    col1[i] = normal_train_data[i][0]
    col2[i] = normal_train_data[i][1]
train_2 = np.column_stack([col1,col2])
clf = Perceptron(max_iter = 1000, tol = 1e-15, random_state = None)
clf.fit(train_2, class_label)
print('\nFinal 3 Weight vectors for first 2 features')
print(clf.coef_)
predict_train = clf.predict(train_2)
train_2_accuracy = accuracy_score(class_label, predict_train)
print("Training data accuracy for first 2 features = ",train_2_accuracy)

#Test data
filename_1 = 'wine_test.csv'
dataset_1 = ld_csv(filename_1)
features_test = copy.deepcopy(dataset_1)
for row in features_test:
    del row[-1]
test_data = np.array(features_test).astype("float")
label_test = []
for row in dataset_1:
    label_test.append(row[-1])
class_label_1 = np.array(label_test).astype("float")
normal_test_data = scaler.transform(test_data)
colm1 = np.zeros(len(normal_test_data))
colm2 = np.zeros(len(normal_test_data))
for i in range(len(normal_test_data)):
    colm1[i] = normal_test_data[i][0]
    colm2[i] = normal_test_data[i][1]
test_2 = np.column_stack([colm1,colm2])
predict_test = clf.predict(test_2)
test_2_accuracy = accuracy_score(class_label_1, predict_test)
print("Test data accuracy for first 2 features = ",test_2_accuracy)


#Perceptron classifier for all 13 features
#Training data
clf1 = Perceptron(max_iter = 1000, tol = 1e-15, random_state = None)
clf1.fit(normal_train_data, class_label)
print('\nFinal 3 Weight vectors for all 13 features')
print(clf1.coef_)
predict_train_13 = clf1.predict(normal_train_data)
train_13_accuracy = accuracy_score(class_label, predict_train_13)
print("Training data accuracy for all 13 features = ", train_13_accuracy)

#Test data
predict_test_13 = clf1.predict(normal_test_data)
test_13_accuracy = accuracy_score(class_label_1, predict_test_13)
print("Test data accuracy for all 13 features = ", test_13_accuracy)


#Best performance model
#Perceptron for first two features
best_wt2 = np.zeros((3,2))
lst = np.zeros((3,2))
ini_wt = np.zeros((3,2))
best_accuracy = 0
for i in range(100):
    clf2 = Perceptron(max_iter = 1000, tol = None, random_state = None)
    lst = np.random.randn(3,2)
    clf2.fit(train_2, class_label, coef_init = lst)
    train_pred_2 = clf2.predict(train_2)
    train_2_accuracy = accuracy_score(class_label, train_pred_2)
    if(train_2_accuracy > best_accuracy):
        best_accuracy = train_2_accuracy
        best_wt2 = clf2.coef_
        ini_wt = lst
print("\nBest final weight vectors for first 2 features\n", best_wt2)
clf2.fit(train_2, class_label, coef_init = ini_wt)
print("Best Training data accuracy for first 2 features = ",best_accuracy)
predict_test2 = clf2.predict(test_2)
test_2_accuracy = accuracy_score(class_label_1, predict_test2)
print("Best Test data accuracy for first 2 features = ",test_2_accuracy)

#Perceptron for all 13 features
best_wt13 = np.zeros((3,13))
lst13 = np.zeros((3,13))
ini_wt13 = np.zeros((3,13))
best_accuracy = 0
for i in range(100):
    clf3 = Perceptron(max_iter=1000, tol=None, random_state = None)
    lst13 = np.random.randn(3,13)
    clf3.fit(normal_train_data, class_label, coef_init = lst13)
    train_pred_13 = clf3.predict(normal_train_data)
    train_13_accuracy = accuracy_score(class_label, train_pred_13)
    if(train_13_accuracy > best_accuracy):
        best_accuracy = train_13_accuracy
        best_wt13 = clf3.coef_
        ini_wt13 = lst13
print("\nBest final weight vectors for all 13 features\n", best_wt13)
clf3.fit(normal_train_data, class_label, coef_init = ini_wt13)
print("Best Training data accuracy for all 13 features = ",best_accuracy)
predict_test13 = clf3.predict(normal_test_data)
test_2_accuracy = accuracy_score(class_label_1, predict_test13)
print("Best test data accuracy for all 13 features = ",test_2_accuracy)
```

# MSE Classifier

```python
#MSE Classifier

from sklearn.linear_model import LinearRegression
from sklearn.multiclass import OneVsRestClassifier

class MSE_binary(LinearRegression):
    def __init__(self):
        super(MSE_binary, self).__init__()
    def predict(self, X):
        thr = 0.5
        y = self._decision_function(X)
        y_int = np.zeros(len(X))
        for i in range(len(X)):
            if(y[i]<thr):
                y_int[i] = 0
            else:
                y_int[i] = 1
        return y_int
    
feat1_train = np.zeros(len(train_data))
feat2_train = np.zeros(len(train_data))
for i in range(len(train_data)):
    feat1_train[i] = train_data[i][0]
    feat2_train[i] = train_data[i][1]
train_unnorm = np.column_stack([feat1_train,feat2_train])

feat1_test = np.zeros(len(test_data))
feat2_test = np.zeros(len(test_data))
for i in range(len(test_data)):
    feat1_test[i] = test_data[i][0]
    feat2_test[i] = test_data[i][1]
test_unnorm = np.column_stack([feat1_test,feat2_test])

#MSE classifier for first two features for unnormalised data
binary_model = MSE_binary()
model = OneVsRestClassifier(binary_model)
model.fit(train_unnorm, class_label)
pred_unnorm = model.predict(test_unnorm)
accuracy_unnorm2 = accuracy_score(class_label_1, pred_unnorm)
print("Test data accuracy for first 2 features for unnormalized data = ", accuracy_unnorm2)

#MSE classifier for all 13 features for unnormalised data
binary_model = MSE_binary()
model = OneVsRestClassifier(binary_model)
model.fit(train_data, class_label)
pred_unnorm = model.predict(test_data)
accuracy_unnorm13 = accuracy_score(class_label_1, pred_unnorm)
print("Test data accuracy for all 13 features for unnormalized data = ", accuracy_unnorm13)


#MSE classifier for first two features for standardized data
binary_model = MSE_binary()
model = OneVsRestClassifier(binary_model)
model.fit(train_2, class_label)
pred_std = model.predict(test_2)
accuracy_std2 = accuracy_score(class_label_1, pred_std)
print("\nTest data accuracy for first 2 features for Standardized data = ",accuracy_std2)

#MSE classifier for all 13 features for standardized data
#Training data
binary_model = MSE_binary()
model = OneVsRestClassifier(binary_model)
model.fit(normal_train_data, class_label)
pred_std = model.predict(normal_test_data)
accuracy_std13 = accuracy_score(class_label_1, pred_std)
print("Test data accuracy for all 13 features for Standardized data = ",accuracy_std13)
```
